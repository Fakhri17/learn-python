```python
tabel = []
for i in daftar:
    tabel.append(tuple(i.split(" - ")))

nama = [i[0] for i in tabel]
nim = [i[1] for i in tabel]
kota = [i[2].upper() for i in tabel]
gender = [i[3] for i in tabel]
nilaiUJIAN = np.random.randint(40,101,43) 
nilaiTUGAS = np.random.randint(40,101,43)
nilaiKUIS = np.random.randint(40,101,43)

data = pd.DataFrame({"nama": nama, "nim": nim, "gender": gender, "kota": kota, "Nilai Kuis": nilaiKUIS, "Nilai Tugas": nilaiTUGAS, "Nilai Ujian": nilaiUJIAN })
data.columns = ["Nama", "NIM", "Jenis Kelamin", "Asal Kota", "Nilai Kuis", "Nilai Tugas", "Nilai Ujian"]
data["Total Nilai"]= data["Nilai Kuis"]*0.3+data["Nilai Tugas"]*0.3+data["Nilai Ujian"]*0.4
data
```

```python
data = data[data["Jenis Kelamin"] == "L"]

data = data.head(10)
data
```
